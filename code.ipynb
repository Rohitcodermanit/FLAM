```python
# Parameters
theta = np.pi / 6   # 30 degrees
M = 0.03
X = 55.185

# t range
t = np.linspace(6, 60, 1000)


# Equations
x = t * np.cos(theta) - np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.sin(theta) + X
y = 42 + t * np.sin(theta) + np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.cos(theta)

# Plot
plt.figure(figsize=(8,6))
plt.plot(x, y, color='teal', linewidth=2,label='eqaution')
sns.lineplot(x=data['x'], y=data['y'],label='data_point')
plt.legend()
plt.title("Parametric Curve v/s Real curve")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
